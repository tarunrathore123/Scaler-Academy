```python
CoPrime Again

Problem Description
You are given 2 arrays A and B of size N and M respectively and a number K. 
You have to tell that how many pairs (A[i], B[j]) (1 <= i <= N and 1 <= j <= m) exists 
such that product of them is not CoPrime with K.   


Problem Constraints
1 <= N, M <= 1e5
1 <= A[i], B[j], K <= 1e6


Input Format
Given two Arrays A and B of Type Integer, and an Integer K.


Output Format
Return a single Integer P, the number of pairs.


Example Input
 A = [1, 2, 3]
 B = [2, 3, 4, 5] 
 K = 3


Example Output
 6


Example Explanation
 There are total 12 pairs, and 6 pairs are there such that their product is not 
    coprime with k, i.e 6.
 These Pairs are :
    (1, 3), (2, 3), (3, 3), (3, 2), (3, 4), (3, 5)
```

```python
# Tip: Recursive solution take much more time than an iterative solution. 
# So, prefer iterative over recursive

class Solution:
    # @param A : list of integers
    # @param B : list of integers
    # @param C : integer
    # @return an integer
    
    def gcd(self, a, b):
        
        while(b):
            a,b = b,a%b
        return a
        

    
    def solve(self, A, B, C):
        
        m = len(A)
        n = len(B)
        cnt1= 0
        cnt2 = 0
        for i in range(m):
            if self.gcd(A[i],C)==1:
                cnt1 += 1
        
        for j in range(n):
            if self.gcd(B[j],C)==1:
                cnt2 += 1
        tot = m*n
        cop = cnt1*cnt2
        return tot-cop
        
            
# def gcd(x,y):
#     while(y):
#         x,y=y,x%y
#     return x    

# class Solution:
    
#     def solve(self, A, B, C):
#         n=len(A)
#         m=len(B)
#         ca=0
#         cb=0
#         for i  in range(n):
#             if(gcd(A[i],C)==1):
#                 ca+=1
#         for i in range(m):
#             if(gcd(B[i],C)==1):
#                 cb+=1
#         tot=n*m
#         cp=ca*cb
#         return tot-cp
        
```
